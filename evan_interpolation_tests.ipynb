```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interp
```

```python
# 1-dimensional "center of mass" calculation
pts = np.zeros(2)
n = pts.shape[0]
for i in range(n):
    pts[i] = np.random.rand()
center_1d = (-0.5*pts[0] + 0.5*pts[1])/np.sum(pts)
print("Point 1:", pts[0], "\nPoint 2:", pts[1], "\nAverage:", center_1d)
```

```python
# extending to 2-dimensional CoM
# 0 1
# 2 3
pts_2d = np.zeros(4)
n = pts_2d.shape[0]
# for i in range(n):
    # pts_2d[i] = np.random.rand()
pts_2d = [0.1, 0.1, 0.1, 0.8]
center_x = (-0.5*(pts_2d[0] + pts_2d[2]) + 0.5*(pts_2d[1] + pts_2d[3]))/np.sum(pts_2d) + 0.5
center_y = (-0.5*(pts_2d[2] + pts_2d[3]) + 0.5*(pts_2d[0] + pts_2d[1]))/np.sum(pts_2d) + 0.5 # add for normalization
print(pts_2d)
print("x CoM:", center_x, "y CoM:", center_y)
plt.plot(center_x, center_y, "ro")
plt.xlim(xmin=0, xmax=1)
plt.ylim(ymin=0, ymax=1)
```
